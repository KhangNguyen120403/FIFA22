```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import statsmodels.formula.api as smf
```

```python
#2a
results = smf.ols('rank ~ passing + attacking + defending + skill', data=df).fit()
print(results.summary())
```

```python

Y = df[["rank"]]
X = df[["passing", "attacking", "defending", "skill"]]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=2023)
X_test
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, color='blue')
plt.xlabel("predicted value") 
plt.ylabel("Residuals")
plt.title("predicted value vs Residuals")
```

2b: The problem is that the residuals are all in 1 area, and they are not spread out randomly across the scatterplot.

#2c 70.5% of the variation in rank can be explained by the features, because R-squared is 0.705

2d A 1-unit increase in skill is associated with 0.0066-unit increase in ranking, holding passing, attacking, and defending constant.

```python
#2e
confidence_interval = results.conf_int(alpha=0.05).loc['skill']
print("This is the confidence interval:", confidence_interval)
```

#2f The confidence interval indicates the range within which the true coefficient for "skill" is likely to fall, with 95% confidence. The confidence interval for "skill" is [-0.011086,0.024249] , this suggests that, holding all other variables constant (passing, attacking, defending), a one-unit increase in "skill" is associated with a change in ranking by between -0.011086 to 0.024249 points.

```python
#4e
X = df[['shooting', 'passing', 'dribbling', 'defending', 'attacking', 'skill', 'movement', 'power', 'mentality','goalkeeping']]
Y = df[['preferred_foot']]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=456)
X_train.head()
```

```python
X_train
```

```python
y_train
```

```python
#4f

accuracy = []
kValues = range(1,31)

for k in kValues:
    knn = KNeighborsClassifier(n_neighbors = k)

    knn.fit(X_train, y_train)

    y_pred = knn.predict(X_test)

    accuracy.append(accuracy_score(y_test, y_pred))


plt.plot(kValues, accuracy)
plt.title('Accuracy vs. Number of Neighbors (k)')
plt.xlabel('K Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
#4g
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
y_pred[:3]
```

```python
#4h
confusion_matrix = confusion_matrix(y_test, y_pred)
confusion_matrix
```

1132 players who prefer their left foot were predicted to prefer their right foot.

```python
#4i
print(classification_report(y_test, y_pred))
```

The recall for the classification "Left" in classification report is 0.15, which suggests that out of all actual "Left" foot players in the test set, the model is correctly identifying only 15% of them. It's failing to detect most of the positive cases for the "Left" foot player.

#4j
With an accuracy of 73%, the model is correct in its predictions for the majority of the observations. However, the model exhibits a significant bias towards the right foot players, as indicated by the higher recall for "Right" (0.93) compared to "Left" (0.15). The precision for left foot players is also low (0.42), which means that even when the model predicts a player to be left foot, there's a considerable chance it's incorrect. So the model predicts right foot players with high accuracy, it does a poor job of identifying left foot players.
